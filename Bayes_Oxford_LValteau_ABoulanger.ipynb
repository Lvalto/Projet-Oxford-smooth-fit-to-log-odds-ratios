# Initialisation des paramètres

```python
alpha_IN = 0
beta1_IN = 0
beta2_IN = 0
sigma_IN = 1

#Mu_i et b_i sont de taille 120x1
mu_IN = np.zeros(K)
b_IN  = np.zeros(K)
```

# Algorithme Metropolis-Hastings within Gibbs

```python
# L'objectif de la fonction MHG est d'utiliser "M.-H. within Gibbs sampler" slide 51/85 pour les paramètres alpha, beta1, beta2 et sigma**2

# n0, n1, r0, r1 --> Données
# nchain ----------> taille de la chaine
# K ---------------> taille des Données
# Parameter_IN ----> initialisation des paramètres
# prop_sd ---------> variances des marches aléatoires des propositons pour chaqe paramètres


def MHG(nchain, alpha_IN, beta1_IN, beta2_IN, sigma_IN, mu_IN, b_IN, prop_sd, K, r1, n1, r0, n0, year):
    
    #Initialisation des paramètres
    alpha = alpha_IN
    beta1 = beta1_IN
    beta2 = beta2_IN
    sigma = sigma_IN
    mu = mu_IN
    b = b_IN
    log_psi = alpha + beta1 * year + beta2 * (year**2 - 22) + b
    p0 = expit(mu)
    p1 = expit(mu + log_psi)
    
    #Chaines différentes car les tailles ne sont pas les mêmes
    chain = np.zeros((nchain+1, 4)) # chaine de alpha, beta1, beta2 et sigma **2
    chain_b = np.zeros((nchain+1, K)) # chaine de b
    chain_mu = np.zeros((nchain+1, K)) # chaine de mu
    
    # Initialisation de la chaîne
    chain[0,0] = alpha_IN
    chain[0,1] = beta1_IN
    chain[0,2] = beta2_IN
    chain[0,3] = sigma_IN
    chain_b[0,:] = b_IN
    chain_mu[0,:] = mu_IN

    #Initialisation des taux d'acceptation
    acc_rate_all = np.zeros(3)
    acc_rate_b = np.zeros(K)
    acc_rate_mu = np.zeros(K)    
    
    for i in range(nchain):
        
        # Mise à jour de alpha
        prop = normal(loc = alpha, scale = prop_sd[0])
        prop_log_psi = prop + beta1 * year + beta2 * (year**2 - 22) + b
        prop_p1 = expit(mu + prop_log_psi)
        
        top = - (prop**2 / (2 * 1e6)) + sum(r1 * np.log(prop_p1)) + sum((n1-r1) * np.log(1 - prop_p1))
        bottom = - (alpha**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            alpha = prop
            log_psi = prop_log_psi
            p1 = prop_p1
            acc_rate_all[0] += 1
        
        # Mise à jour de beta1
        prop = normal(loc = beta1, scale = prop_sd[1])
        prop_log_psi = alpha + prop * year + beta2 * (year**2 - 22) + b
        prop_p1 = expit(mu + prop_log_psi)
        
        top = - (prop**2 / (2 * 1e6)) + sum(r1 * np.log(prop_p1)) + sum((n1-r1) * np.log(1 - prop_p1))
        bottom = - (beta1**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta1 = prop
            log_psi = prop_log_psi
            p1 = prop_p1
            acc_rate_all[1] += 1
            
        # Mise à jour de beta2
        prop = normal(loc = beta2, scale = prop_sd[2])
        prop_log_psi = alpha + beta1 * year + prop * (year**2 - 22) + b
        prop_p1 = expit(mu + prop_log_psi)
        
        top = - (prop**2 / (2 * 1e6)) + sum(r1 * np.log(prop_p1)) + sum((n1-r1) * np.log(1 - prop_p1))
        bottom = - (beta2**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta2 = prop
            log_psi = prop_log_psi
            p1 = prop_p1
            acc_rate_all[2] += 1
            
        # Mise à jour de sigma -> Loi conjuguée
        sigma = 1 / gamma(shape = 1e-3 + K/2, scale = 1e-3 + 0.5 * sum(b**2))
            
        for j in range(K):
            
            # Mise à jour de mu
            prop = normal(loc = mu[j], scale = prop_sd[3])
            prop_p0_j = expit(prop)
            prop_p1_j = expit(prop + log_psi[j])
            
            top = - (prop**2 / (2 * 1e6)) + r0[j] * np.log(prop_p0_j) + (n0[j] - r0[j]) * np.log(1 - prop_p0_j) + r1[j] * np.log(prop_p1_j) + (n1[j] - r1[j]) * np.log(1 - prop_p1_j)
            bottom = - (mu[j]**2 / (2 * 1e6)) + r0[j] * np.log(p0[j]) + (n0[j] - r0[j]) * np.log(1 - p0[j]) + r1[j] * np.log(p1[j]) + (n1[j] - r1[j]) * np.log(1 - p1[j])
            
            acc_prob = np.exp(top - bottom)
            
            if np.random.uniform() < acc_prob:
                mu[j] = prop
                p0[j] = prop_p0_j
                p1[j] = prop_p1_j
                acc_rate_mu[j] += 1
                
            # Mise à jour de b
            prop = normal(loc = b[j], scale = prop_sd[4])
            prop_log_psi_j = alpha + beta1 * year[j] + beta2 * (year[j]**2 - 22) + prop
            prop_p1_j = expit(mu[j] + prop_log_psi_j)
            
            top = - (prop**2 / (2 * sigma)) + r1[j] * np.log(prop_p1_j) + (n1[j] - r1[j]) * np.log(1 - prop_p1_j)
            bottom = - (b[j]**2 / (2 * sigma)) + r1[j] * np.log(p1[j]) + (n1[j] - r1[j]) * np.log(1 - p1[j])
            
            acc_prob = np.exp(top - bottom)
            
            if np.random.uniform() < acc_prob:
                b[j] = prop
                log_psi[j] = prop_log_psi_j
                p1[j] = prop_p1_j
                acc_rate_b[j] += 1
            
        # Mise à jour des chaines
        chain[i+1,:] = np.array([alpha, beta1, beta2, sigma])
        chain_b[i+1,:] = b 
        chain_mu[i+1,:] = mu
        
    return(chain, chain_mu, chain_b, acc_rate_all, acc_rate_mu, acc_rate_b)  
```

```python
chain, b_chain, mu_chain, acc_rate, b_acc_rate, mu_acc_rate = MHG(10000, alpha_IN, beta1_IN, beta2_IN, sigma_IN, mu_IN, b_IN, np.array([0.05, 0.01, 0.003, 0.1, 0.5]), K, r1, n1, r0, n0, year)
```
